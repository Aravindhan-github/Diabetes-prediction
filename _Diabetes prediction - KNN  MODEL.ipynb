```python
x=df.drop("Outcome",axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.isnull().sum()
```

```python
#Handling Missing value  for x_train
Glucose_mean=x_train['Glucose'].mean()
BloodPressure_mean=x_train['BloodPressure'].mean()
```

```python
skin_Thickness_median=x_train['SkinThickness'].median()
Insulin_median = x_train['Insulin'].median()
BMI_median = x_train['BMI'].median()
```

```python
 x_train['Glucose'].fillna(Glucose_mean,inplace=True)
```

```python
 x_train['BloodPressure'].fillna(BloodPressure_mean,inplace=True)
```

```python
x_train['SkinThickness'].fillna(skin_Thickness_median,inplace=True)
```

```python
x_train['Insulin'].fillna(Insulin_median,inplace=True)
```

```python
x_train['BMI'].fillna(BMI_median,inplace=True)
```

```python
x_train.isnull().sum()
```

```python
#handling missing value for x_test
x_test['Glucose'].fillna(Glucose_mean,inplace=True)
x_test['BloodPressure'].fillna(BloodPressure_mean,inplace=True)
x_test['SkinThickness'].fillna(skin_Thickness_median,inplace=True)
x_test['Insulin'].fillna(Insulin_median,inplace=True)
x_test['BMI'].fillna(BMI_median,inplace=True)
```

```python
x_test.isnull().sum()
```

```python
x_test
```

```python
x_train
```

```python
#Feature Scaling using Normalization
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
```

```python
x_train=min_max_scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=min_max_scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=10)
```

```python
knn_model.fit(x_train,y_train)
```

```python
y_predict=knn_model.predict(x_test)
```

```python
#Model Evaluation
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
